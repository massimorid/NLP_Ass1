```python
#list of bigrams
bigramms = ngrams(text.split(), 2)

bigramsList = []
for x in bigramms:
  bigramsList.append(x)
```

```python
#showing the content of the list
bigramsList
```

```python
#from_bigram_to_next_token_counts dictionary
from_bigram_to_next_token_counts = {}

for i in range(len(bigramsList) - 1):

    #if the element is not present in the dictionary initialize it with a sub dictionary
    if bigramsList[i] not in from_bigram_to_next_token_counts:
        from_bigram_to_next_token_counts[bigramsList[i]] = {}

    #add the count of the following word
    from_bigram_to_next_token_counts[bigramsList[i]][bigramsList[i + 1][1]] = from_bigram_to_next_token_counts[bigramsList[i]].get(bigramsList[i + 1][1], 0) + 1
```

```python
#showing result
from_bigram_to_next_token_counts
```

```python
for subdic in from_bigram_to_next_token_counts.values():
  #for every subdictionary calculate the total value
    total = sum(subdic.values())
    #divide every element in the subdictionary by the total to give it a weight
    for key in subdic:
        subdic[key] /= total
```

```python
from_bigram_to_next_token_probs = from_bigram_to_next_token_counts
```

###Part 3

```python
def sample_next_token(w1, w2):
  """ The function takes in two words that compose the bigram, and outputs a predicted word"""


  bigram = (w1, w2)

  #checks existance of the bigram
  if bigram in from_bigram_to_next_token_probs:

    next_word_probs = from_bigram_to_next_token_probs[bigram]
    words = list(next_word_probs.keys())
    probabilities = list(next_word_probs.values())

    #random weighted samples the next word
    return random.choices(words, weights=probabilities, k=1)[0]


  #There could be a chance that the bigram is not existant
  else:
    print("No word prediction available based on text provided")
```

```python
#example
sample_next_token('lady', 'macbeth')
```

```python
def generate_text_from_bigram(w1, w2, wGenerated):

  """iterative solution to output the next word generated"""
  bigram = (w1, w2)

  #checks if terminating condition is met
  if wGenerated != 0:

    #uses previous function
    nextword = sample_next_token(w1, w2)
    print(nextword)

    #iterative call of the function
    generate_text_from_bigram(w2, nextword, wGenerated-1)

  return 0
```

```python

from collections import defaultdict
import random

# Trigram counts and probabilities
def from_trigram_to_next_token_counts(corpus):
    trigram_counts = defaultdict(lambda: defaultdict(int))
    for sentence in corpus:
        for i in range(len(sentence) - 2):
            trigram = (sentence[i], sentence[i + 1], sentence[i + 2])
            trigram_counts[trigram[:-1]][trigram[-1]] += 1
    return trigram_counts

def from_trigram_to_next_token_probs(trigram_counts):
    trigram_probs = {}
    for trigram, next_words in trigram_counts.items():
        total = sum(next_words.values())
        trigram_probs[trigram] = {word: count / total for word, count in next_words.items()}
    return trigram_probs

# Quadgram counts and probabilities
def from_quadgram_to_next_token_counts(corpus):
    quadgram_counts = defaultdict(lambda: defaultdict(int))
    for sentence in corpus:
        for i in range(len(sentence) - 3):
            quadgram = (sentence[i], sentence[i + 1], sentence[i + 2], sentence[i + 3])
            quadgram_counts[quadgram[:-1]][quadgram[-1]] += 1
    return quadgram_counts

def from_quadgram_to_next_token_probs(quadgram_counts):
    quadgram_probs = {}
    for quadgram, next_words in quadgram_counts.items():
        total = sum(next_words.values())
        quadgram_probs[quadgram] = {word: count / total for word, count in next_words.items()}
    return quadgram_probs

# Sampling functions for trigram and quadgram
def sample_next_token_from_trigram(w1, w2, trigram_probs):
    trigram = (w1, w2)
    if trigram in trigram_probs:
        next_word_probs = trigram_probs[trigram]
        words = list(next_word_probs.keys())
        probabilities = list(next_word_probs.values())
        return random.choices(words, weights=probabilities, k=1)[0]
    else:
        print("No word prediction available for the trigram.")
        return None

def sample_next_token_from_quadgram(w1, w2, w3, quadgram_probs):
    quadgram = (w1, w2, w3)
    if quadgram in quadgram_probs:
        next_word_probs = quadgram_probs[quadgram]
        words = list(next_word_probs.keys())
        probabilities = list(next_word_probs.values())
        return random.choices(words, weights=probabilities, k=1)[0]
    else:
        print("No word prediction available for the quadgram.")
        return None

# Text generation functions for trigram and quadgram
def generate_text_from_trigram(w1, w2, n, trigram_probs):
    if n > 0:
        next_word = sample_next_token_from_trigram(w1, w2, trigram_probs)
        if next_word:
            print(next_word, end=" ")
            generate_text_from_trigram(w2, next_word, n - 1, trigram_probs)

def generate_text_from_quadgram(w1, w2, w3, n, quadgram_probs):
    if n > 0:
        next_word = sample_next_token_from_quadgram(w1, w2, w3, quadgram_probs)
        if next_word:
            print(next_word, end=" ")
            generate_text_from_quadgram(w2, w3, next_word, n - 1, quadgram_probs)
```
